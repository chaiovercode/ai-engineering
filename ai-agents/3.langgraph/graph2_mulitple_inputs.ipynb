```python
from typing import TypedDict, Dict, List
from langgraph.graph import StateGraph
```

```python
# you can name the class whatever you want but AgentState is more intutive
# and explains the purpose of the class

class AgentState(TypedDict):
    values: List[int]
    name: str
    result: str
```

```python
def process_values(state: AgentState) -> AgentState:
    """Process the values in the state."""
    state['result'] = f"Hi there {state['name']}. The sum of your values is {sum(state['values'])}."
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("process_values", process_values)
graph.set_entry_point("process_values")
graph.set_finish_point("process_values")

app = graph.compile()

from IPython.display import display, Image
display(Image(app.get_graph().draw_mermaid_png()))

```

```python
result = app.invoke({"values":[1,2,3,4], "name":"Alice"})

result=(result["result"])

print(result)


```
